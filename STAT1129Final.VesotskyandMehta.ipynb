```python
import turtle
import time
from matplotlib import colors

scr = turtle.getscreen() 
scr.title("USA Flag")
scr.bgcolor("white")


t = turtle.Turtle()
t.speed(20)
t.penup()

# flag height and width
flag_ht = 250
flag_lth = 475

# where flag starts on the screen 
x1 = -250
y1 = 120

# defining sizes of stars and stripes
stripe_ht = flag_ht/13
stripe_lth = flag_lth
star_size = 12

def get_color(string):
    return colors.to_rgb(string) 

def draw_rectangle(x, y, height, length, color):
    t.goto(x,y)
    t.pendown()
    t.color(color)
    t.begin_fill()
    t.forward(length)
    t.right(90)
    t.forward(height)
    t.right(90) 
    t.forward(length)
    t.right(90)
    t.forward(height)
    t.right(90)
    t.end_fill()
    t.penup() 

def draw_star(x,y,color,size):
    t.goto(x,y)
    t.setheading(0)
    t.pendown()
    t.begin_fill()
    t.color(color)
    for turn in range(0,5):
        t.forward(size)
        t.right(144)
        t.forward(size)
        t.right(144)
    t.end_fill()
    t.penup()

def draw_stripes():
    x = x1
    y = y1
    for stripe in range(0,6):
        for color in ["red", "white"]:
            draw_rectangle(x, y, stripe_ht, stripe_lth, color)
            y = y - stripe_ht

    draw_rectangle(x, y, stripe_ht, stripe_lth, 'red')
    y = y - stripe_ht

def draw_square():
    square_ht = (7/13) * flag_ht
    square_wdt = (0.76) * flag_ht
    draw_rectangle(x1, y1, square_ht, square_wdt, 'navy')

def stars_6row():
    dist_of_stars = 30
    dist_bet_lines = stripe_ht + 6
    y = 112
    for row in range(0,5):
        x = -234
        for star in range (0,6):
            draw_star(x, y, 'white', star_size)
            x = x + dist_of_stars
        y = y - dist_bet_lines

def stars_5row():
    dist_of_stars = 30
    dist_between_lines = stripe_ht + 6
    y = 100
    for row in range(0,4):
        x = -217
        for star in range (0,5):
            draw_star(x, y, 'white', star_size)
            x = x + dist_of_stars
        y = y - dist_between_lines
    
def draw_flag():
    draw_rectangle(-250, 120, 250, 475, "white")
    draw_stripes()
    draw_square()
    stars_6row()
    stars_5row()

def main():
    draw_flag()    

time.sleep(1)
t.hideturtle()

if __name__ == "__main__":
    main()

scr.mainloop
```
