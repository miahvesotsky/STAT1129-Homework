```python
import turtle
import time
from matplotlib import colors

scr = turtle.getscreen() 
scr.title("USA Flag")
scr.bgcolor("white")


t = turtle.Turtle()
t.speed(20)
t.penup()

def get_color(string):
    return colors.to_rgb(string) 

def draw_rectangle(x, y, height, length, color):
    t.goto(x,y)
    t.pendown()
    t.color(color)
    t.begin_fill()
    t.forward(length)
    t.right(90)
    t.forward(height)
    t.right(90) 
    t.forward(length)
    t.right(90)
    t.forward(height)
    t.right(90)
    t.end_fill()
    t.penup() 

def draw_star(x,y,color,size):
    t.goto(x,y)
    t.setheading(0)
    t.pendown()
    t.begin_fill()
    t.color(color)
    for turn in range(0,5):
        t.forward(size)
        t.right(144)
        t.forward(size)
        t.right(144)
    t.end_fill()
    t.penup()

def draw_stripes():
    x = -250
    y = 120
    for stripe in range(0,6):
        for color in ["#B31942", "white"]:
            draw_rectangle(x, y, (250/13), 475, color)
            y = y - (250/13)

    draw_rectangle(x, y, (250/13), 475, '#B31942')
    y = y - (250/13)

def draw_square():
    square_ht = (7/13) * 250
    square_wdt = (0.76) * 250
    draw_rectangle(-250, 120, square_ht, square_wdt, '#0A3161')

def stars_6row():
    dist_of_stars = 30
    dist_bet_lines = (250/13) + 6
    y = 112
    for row in range(0,5):
        x = -234
        for star in range (0,6):
            draw_star(x, y, 'white', 12)
            x = x + dist_of_stars
        y = y - dist_bet_lines

def stars_5row():
    dist_of_stars = 30
    dist_between_lines = (250/13) + 6
    y = 100
    for row in range(0,4):
        x = -217
        for star in range (0,5):
            draw_star(x, y, 'white', 12)
            x = x + dist_of_stars
        y = y - dist_between_lines
    
def draw_flag():
    draw_rectangle(-250, 120, 250, 475, "white")
    draw_stripes()
    draw_square()
    stars_6row()
    stars_5row() 

def main():
    draw_flag()    

time.sleep(1)
t.hideturtle()

if __name__ == "__main__":
    main()

scr.mainloop
```
