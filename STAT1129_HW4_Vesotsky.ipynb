```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([88, 66, 90, 55, 77])
x = np.array(['Andy', 'Amy', 'James', 'Jules', 'Arthur'])

plt.title('Student Grades')
plt.pie(y, labels = x)
plt.legend(bbox_to_anchor = (1.0, 1.0), loc = 'upper left')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#plot1
x1 = np.array([0, 1, 2, 3, 4])
y1 = np.array([1, 2, 3, 4, 5])
x2 = np.array([0, 1, 2, 3, 4])
y2 = np.array([5, 6, 7, 8, 9])
x3 = np.array([0, 1, 2, 3, 4])
y3 = np.array([9, 10, 11, 12, 13])

plt.subplot(2, 3, 1)
plt.plot(x1, y1, x2, y2, x3, y3)
plt.title('My Lines')

#plot2
x = np.array(['A', 'B', 'C'])
y = np.array([3, 5, 7])

plt.subplot(2, 3, 2)
plt.bar(x, y)
plt.title('My Bar')

#plot3
y = np.array([3, 5, 7])
x = np.array(['apple', 'banana', 'cherry'])

plt.subplot(2, 3, 3)
plt.pie(y, labels = x)
plt.title('My Fruit Pie')

#plot4
x = np.array([2, 3, 5, 8])
y = np.array([5, 6, 6, 3])

plt.subplot(2, 3, 4)
plt.scatter(x, y)
plt.title('My Scatter')

#plot5
x = np.random.normal([5, 6, 7])

plt.subplot(2, 3, 5)
plt.hist(x)
plt.title('My Histogram')

#plot6
x = [(2, 4, 6, 8)]
y = [(2, 4, 6, 8)]

plt.subplot(2, 3, 6)
plt.plot(x, y)
plt.grid(color = 'blue', linestyle = '-', linewidth = 0.5)
plt.title('My Grid')

plt.show()
```
